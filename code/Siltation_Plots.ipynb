```python
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Sample data (replace this with your actual data)
siltation_data = [10, 15, 20, 25, 18, 22, 28, 30]
start_time = datetime(2023, 12, 20, 9, 0, 0)

# Generate datetime values for x-axis
date_times = [start_time + timedelta(hours=i) for i in range(len(siltation_data))]

# Plotting the graph
plt.plot(date_times, siltation_data, marker='o')
plt.title('Siltation Level Over Time')
plt.xlabel('DateTime')
plt.ylabel('Siltation Level')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility

# Display the plot
plt.tight_layout()
plt.show()
```
